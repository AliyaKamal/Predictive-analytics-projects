```python
#(1)   Random Forest

#RF = RandomForestRegressor(random_state=616).fit(train_X, train_y)
#predicted_y_training = RF.predict(train_X)
from sklearn.metrics import mean_squared_error

RF = RandomForestRegressor(n_estimators=100, random_state=616).fit(train_X, train_y)
#performance training
predicted_y_training = RF.predict(train_X)
mse = mean_squared_error(train_y, predicted_y_training)
print("Mean Squared Error:", mse)

print("Root Mean Squared Error (RMSE): ", round(mean_squared_error(train_y, predicted_y_training) ** 0.5, 4))
print("Mean Absolute Error (MAE): ", round(mean_absolute_error(train_y,predicted_y_training), 4))
print("Mean Absolute Percentage Error (MAPE): ", round(mean_absolute_percentage_error(train_y, predicted_y_training) * 100, 4))
```

```python
#testing performance
predicted_y_test = RF.predict(test_X)
print(mean_squared_error(test_y, predicted_y_test))

print("Root Mean Squared Error (RMSE): ", round(mean_squared_error(test_y, predicted_y_test) ** 0.5, 4))
print("Mean Absolute Error (MAE): ", round(mean_absolute_error(test_y,predicted_y_test), 4))
print("Mean Absolute Percentage Error (MAPE): ", round(mean_absolute_percentage_error(test_y, predicted_y_test) * 100, 4))
```

```python
feature_imp_RF = pd.Series(RF.feature_importances_, index = predictors.columns)
feature_imp_RF.sort_values(ascending=False)
```

```python
feature_imp_RF.idxmax()
```

```python
# (2)  Gradient Boosting
GB = GradientBoostingRegressor(n_estimators=100, random_state=616).fit(train_X, train_y)
predicted_y_training = GB.predict(train_X)
#training performance

mse = mean_squared_error(train_y, predicted_y_training)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error (RMSE): ", round(mean_squared_error(train_y, predicted_y_training) ** 0.5, 4))
print("Mean Absolute Error (MAE): ", round(mean_absolute_error(train_y,predicted_y_training), 4))
print("Mean Absolute Percentage Error (MAPE): ", round(mean_absolute_percentage_error(train_y, predicted_y_training) * 100, 4))
```

```python
#testing performance
predicted_y_test = GB.predict(test_X)
print(mean_squared_error(test_y, predicted_y_test))

print("Root Mean Squared Error (RMSE): ", round(mean_squared_error(test_y, predicted_y_test) ** 0.5, 4))
print("Mean Absolute Error (MAE): ", round(mean_absolute_error(test_y,predicted_y_test), 4))
print("Mean Absolute Percentage Error (MAPE): ", round(mean_absolute_percentage_error(test_y, predicted_y_test) * 100, 4))
```

```python
feature_imp_GB = pd.Series(GB.feature_importances_, index = predictors.columns)
feature_imp_GB.sort_values(ascending=False)
```

```python
feature_imp_GB.idxmax()
```

```python
# (3) NEURAL NET
#train the NN model and look at performance on train data
# https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html
#verbose=true for progress messages



RN = MLPRegressor(hidden_layer_sizes=(128,128), activation="relu", solver="adam",
                   learning_rate_init=0.01, verbose=True, max_iter=200, batch_size=64).fit(train_X, train_y)
predicted_y_training = RN.predict(train_X)
print(mean_squared_error(train_y, predicted_y_training))

print("Root Mean Squared Error (RMSE): ", round(mean_squared_error(train_y, predicted_y_training) ** 0.5, 4))
print("Mean Absolute Error (MAE): ", round(mean_absolute_error(train_y,predicted_y_training), 4))
print("Mean Absolute Percentage Error (MAPE): ", round(mean_absolute_percentage_error(train_y, predicted_y_training) * 100, 4))
```

```python
# performance on test data

predicted_y_test = RN.predict(test_X)
print(mean_squared_error(test_y, predicted_y_test))

print("Root Mean Squared Error (RMSE): ", round(mean_squared_error(test_y, predicted_y_test) ** 0.5, 4))
print("Mean Absolute Error (MAE): ", round(mean_absolute_error(test_y,predicted_y_test), 4))
print("Mean Absolute Percentage Error (MAPE): ", round(mean_absolute_percentage_error(test_y, predicted_y_test) * 100, 4))
```
